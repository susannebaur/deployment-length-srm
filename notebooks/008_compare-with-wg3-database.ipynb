Just make sure our data actually matches the WG3 database.

```python
%load_ext nb_black
```

```python
import pyam
import scmdata.testing

import utils
```

```python
ar6_wg3_data_dir = utils.DATA_DIR / "raw" / "ar6-wg3"
SOURCE_DATA = str(ar6_wg3_data_dir / "20220314_ar6emissions_harmonized_infilled.csv")

our_set = scmdata.ScmRun(SOURCE_DATA, lowercase_cols=True).filter(variable="*Infilled*")

# Add total CO2 too
our_set_co2_total = our_set.filter(variable="*CO2*").process_over(
    "variable",
    "sum",
    op_cols={"variable": "AR6 climate diagnostics|Infilled|Emissions|CO2"},
    as_run=True,
)
our_set = our_set.append(our_set_co2_total)
our_set.head()
```

```python
wg3_db = pyam.read_iiasa(
    "ar6-public",
    variable=["*|Infilled|*"],
    region="World",
)
wg3_db = scmdata.ScmRun(wg3_db.timeseries())
wg3_db.head()
```

```python
# some data cleaning
wg3_db = wg3_db.filter(unit="unknown", keep=False, log_if_empty=False)
```

Get comparable set

```python
wg3_db["ms"] = wg3_db["model"] + wg3_db["scenario"]
our_set["ms"] = our_set["model"] + our_set["scenario"]

our_set = our_set.filter(ms=wg3_db.get_unique_meta("ms"))
```

```python
our_set.filter(ms=wg3_db.get_unique_meta("ms"))
```

```python
set(wg3_db.get_unique_meta("variable")).intersection(
    set(our_set.get_unique_meta("variable"))
)
```

```python
crossover_variables = set(wg3_db.get_unique_meta("variable")).intersection(
    set(our_set.get_unique_meta("variable"))
)
scmdata.testing.assert_scmdf_almost_equal(
    wg3_db.filter(variable=crossover_variables),
    our_set.filter(variable=crossover_variables, ms=wg3_db.get_unique_meta("ms")),
    check_ts_names=False,
)
```

We can be confident we're using the right timeseries.
